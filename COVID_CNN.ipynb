```python
import numpy as np
import pandas as pd
import os
import tensorflow as tf
import keras
from keras.preprocessing.image import ImageDataGenerator
from keras.preprocessing import image
```

```python
train_datagen = ImageDataGenerator(
        rescale=1./255,
        shear_range=0.2,
        zoom_range=0.2,
        horizontal_flip=True)
training_set = train_datagen.flow_from_directory(
        directory = r"/kaggle/input/chest-xray-pneumonia/chest_xray/chest_xray/train/",
        target_size=(64 , 64),
        batch_size=32,
        class_mode='binary')
```

```python
test_datagen = ImageDataGenerator(rescale=1./255)
test_set = test_datagen.flow_from_directory(
        directory = r"/kaggle/input/chest-xray-pneumonia/chest_xray/chest_xray/test/",
        target_size=(64 , 64),
        batch_size=32,
        class_mode='binary')
```

```python
validation = ImageDataGenerator(rescale=1./255)
vall = test_datagen.flow_from_directory(
        directory = r"/kaggle/input/chest-xray-pneumonia/chest_xray/chest_xray/val/",
        target_size=(64 , 64),
        batch_size=32,
        class_mode='binary')
```

```python
#intiallize CNN
cnn = tf.keras.models.Sequential()
```

```python
#Step1:- Convolution --  rows and columns of feature matrix  -Shape of picture (64,64,Channels=3) -Activation fun
#Convolution2D(numofFiltters,numofRows,numofColumns)
cnn.add(tf.keras.layers.Conv2D(filters = 32 , kernel_size = 3 , activation = 'relu' ))
cnn.add(keras.Input(shape=(64, 64, 3)))
```

```python
#pooling
cnn.add(tf.keras.layers.MaxPool2D( pool_size = 2 , strides = 2))
```

```python
#flattening
cnn.add(tf.keras.layers.Flatten())
```

```python
#full connection
cnn.add(tf.keras.layers.Dense(units = 128 , activation = 'relu'))
```

```python
#adding the output layer
cnn.add(tf.keras.layers.Dense(units = 1 , activation = 'sigmoid'))
```

```python
cnn.compile(optimizer = 'adam', loss = 'binary_crossentropy', metrics = ['accuracy'])
```

```python
cnn.fit(x = training_set , validation_data = test_set , epochs = 25)
```

```python
test_image = image.load_img('../input/chest-xray-pneumonia/chest_xray/test/PNEUMONIA/person109_bacteria_526.jpeg', target_size = (64, 64))
test_image = image.img_to_array(test_image)
test_image = np.expand_dims(test_image, axis = 0)
result = cnn.predict(test_image)
training_set.class_indices
if result[0][0] == 0:
  prediction = 'Negative (normal)'
else:
  prediction = 'Positive (Covid)'

print(prediction)
```
